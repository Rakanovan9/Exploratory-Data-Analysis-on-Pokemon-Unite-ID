```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("PokemonUniteData.csv")
df.head()
```

```python
df.info()
```

```python
df_duplicate = df.duplicated().sum()


df_unique = df.nunique()


print("Amount of duplicates found in data :",df_duplicate,"\n")
print(df_unique)
```

```python

plt.figure(figsize=(10, 6))
df_unique.drop(["Name", "Description"]).plot(kind='bar', color='skyblue')
plt.title("Number of Unique Values in Each Column")
plt.ylabel("Count of Unique Values")
plt.xticks(rotation=45)
plt.show()
```

```python
df_null = df.isnull().sum()
print("Amount of Missing Vales found in data")
df_null
```

```python
df.dtypes
```

```python
df_usagedifficulty_Novice= df[df['UsageDifficulty'] == "Novice"]
print("Amount of Pokemon that has the Usage Difficulty of Novice :",len(df_usagedifficulty_Novice))
df_usagedifficulty_Novice.head()

```

```python
df_usagedifficulty_intermediate = df[df['UsageDifficulty'] == "Intermediate"]
print("Amount of Pokemon that has the Usage Difficulty of Intermediate :",len(df_usagedifficulty_intermediate))
df_usagedifficulty_intermediate.head()
```

```python
df_usagedifficulty_Expert = df[df['UsageDifficulty'] == "Expert"]
print("Amount of Pokemon that has the Usage Difficulty of Expert :",len(df_usagedifficulty_Expert))
df_usagedifficulty_Expert.head()
```

```python
df['Total Stats'] = df[['Offense', 'Endurance', 'Mobility', 'Scoring', 'Support']].sum(axis=1)
df.head()
```

```python
df_melee = df[df['Ranged_or_Melee'] == "Melee"]
df_ranged = df[df['Ranged_or_Melee'] == "Ranged"]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='UsageDifficulty', y='Total Stats', data=df)
plt.title('Comparison of Total Stats by Usage Difficulty')
plt.xlabel('UsageDifficulty')
plt.ylabel('Total Stats')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Role', y='Endurance', data=df)
plt.title('Comparison of Endurance by Usage Difficulty')
plt.xlabel('UsageDifficulty')
plt.ylabel('Role')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Role', y='Mobility', data=df)
plt.title('Comparison of Mobility by Usage Difficulty')
plt.xlabel('UsageDifficulty')
plt.ylabel('Role')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Role', y='Scoring', data=df)
plt.title('Comparison of Scoring by Usage Difficulty')
plt.xlabel('UsageDifficulty')
plt.ylabel('Role')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Role', y='Support', data=df)
plt.title('Comparison of Support by Usage Difficulty')
plt.xlabel('UsageDifficulty')
plt.ylabel('Role')
plt.show()
```

```python
correlation_matrix = df[['Offense', 'Endurance', 'Mobility', 'Scoring', 'Support', 'Total Stats']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)


plt.title('Correlation Matrix')
```

```python
correlation_matrix = df_melee[['Offense', 'Endurance', 'Mobility', 'Scoring', 'Support', 'Total Stats']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)


plt.title('Correlation Matrix Melee')
```

```python
correlation_matrix = df_ranged[['Offense', 'Endurance', 'Mobility', 'Scoring', 'Support', 'Total Stats']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)


plt.title('Correlation Matrix Ranged')
```
